```python
#data[['hr','TLX_mean']].corr()
import numpy as np
# create a list of our conditions
conditions = [
    (data['TLX_mean'] < 10),
    (data['TLX_mean'] >= 10 ) & (data['TLX_mean'] < 20),
    (data['TLX_mean'] >=20)]
values = ['no', 'low', 'medium']
data['Stress'] = np.select(conditions, values)
data
```

```python
conditions1 = [
    (data['TLX_mean'] < 15),
    (data['TLX_mean'] >=15)]
values1 = ['no', 'yes']
data['binar'] = np.select(conditions1, values1)
data
```
